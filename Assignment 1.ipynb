```python
import seaborn as sns
sns.get_dataset_names()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic.drop('deck', axis=1, inplace=True)
```

```python
titanic.info()
```

```python
titanic=titanic.dropna()
```

```python
titanic.shape
```

```python
titanic.describe()
```

```python
titanic[titanic.duplicated()]
```

```python
titanic.skew(numeric_only=True)
```

```python
titanic.kurt(numeric_only=True)
```

```python
titanic.groupby('class',observed=False)[['sex']].value_counts()
```

```python
survived =titanic.groupby('alive',observed=False)[['sex']].value_counts()
```

```python
survived.plot(kind='bar')
plt.title("Survivors by Gender")
plt.xticks(rotation=0) 
```

```python
pclass_dist_by_sex=titanic.groupby('pclass')[['sex']].value_counts()
```

```python
first_class_psngr = titanic[titanic['pclass'] == 1]['embark_town'].value_counts()

first_class_psngr
```

```python
sizes = first_class_psngr.values
labels = ['Southampton', 'Cherbourg', 'Queenstown']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Distribution of First Class Passengers by town")
plt.show()
```

```python
pd.crosstab(titanic['sex'], titanic['pclass']).plot(kind="bar",edgecolor='black')
plt.xticks(rotation=0) 
plt.show()
```

```python
survival_rates = titanic.groupby('pclass')['survived'].mean()# which class passengers survived the most
```

```python
survival_rates.plot(kind='bar', color='skyblue',edgecolor="purple")
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')
plt.xticks(rotation=0)  # Rotate x-axis labels
plt.ylim(0, 1)  # Set y-axis limits to range from 0 to 1
plt.show()
```
