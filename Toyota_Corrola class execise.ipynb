```python
#Calculating the vif values
rsq_Age=smf.ols("Age_08_04~KM+HP+Doors+Gears+Weight", data=df).fit().rsquared
vif_Age=1/(1-rsq_Age)

rsq_KM=smf.ols("KM~Age_08_04+HP+Doors+Gears+Weight", data=df).fit().rsquared
vif_KM=1/(1-rsq_KM)

rsq_HP=smf.ols("HP~Age_08_04+KM+Doors+Gears+Weight", data=df).fit().rsquared
vif_HP=1/(1-rsq_HP)

rsq_D=smf.ols("Doors~Age_08_04+KM+HP+Gears+Weight", data=df).fit().rsquared
vif_D=1/(1-rsq_D)

rsq_G=smf.ols("Gears~Age_08_04+KM+HP+Doors+Weight", data=df).fit().rsquared
vif_G=1/(1-rsq_G)

rsq_W=smf.ols("Weight~Age_08_04+KM+HP+Doors+Gears", data=df).fit().rsquared
vif_W=1/(1-rsq_W)

d1={'Variables':['Age_08_04', 'KM', 'HP', 'Doors', 'Gears', 'Weight'],'VIF':[vif_Age,vif_KM,vif_HP,vif_D,vif_G,vif_W]}
vif_frame=pd.DataFrame(d1)
vif_frame
```
